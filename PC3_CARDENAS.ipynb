```python
import pandas as pd

fallecidos_df = pd.read_csv('fallecidosCOVID.csv')

columnas = ['FECHA_CORTE','FECHA_FALLECIMIENTO','EDAD_DECLARADA','SEXO','CLASIFICACION_DEF','DEPARTAMENTO','PROVINCIA','DISTRITO','UBIGEO','UUID']
fallecidos_df[columnas] = fallecidos_df['FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID'].str.split(';', expand=True)

fallecidos_df = fallecidos_df.drop(columns=['FECHA_CORTE;FECHA_FALLECIMIENTO;EDAD_DECLARADA;SEXO;CLASIFICACION_DEF;DEPARTAMENTO;PROVINCIA;DISTRITO;UBIGEO;UUID'])

fallecidos_df['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_df['FECHA_FALLECIMIENTO'], format='%Y%m%d', errors='coerce')

fallecidos_df['YearMonth'] = fallecidos_df['FECHA_FALLECIMIENTO'].dt.strftime('%Y-%m')

reporte_mensual = pd.pivot_table(fallecidos_df, values='UUID', index='YearMonth', columns='CLASIFICACION_DEF', aggfunc='count', fill_value=0)

reporte_mensual = reporte_mensual.rename(columns={'UUID': 'Fallecidos'})
reporte_mensual=reporte_mensual.iloc[:,[0,1,2,3,4,6]]

reporte_mensual
```
